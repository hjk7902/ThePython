```python
class ZeroLengthError(Exception):
    def __init__(self):
        Exception.__init__(self, 
                           "사용자 정의 예외: 길이가 0임")
```

```python
def insert(data):
    if len(data)==0:
        raise ZeroLengthError()
        
    print(data, "을(를) 입력합니다.", sep='')
```

```python
data = []
try:
    insert(data)
except ZeroLengthError as e:
    print(e)
else:
    print("정상 실행됨")
```
